```python
def ANN(x_train,y_train,x_test,y_test, loss, weights):
    model = keras.Sequential([
    keras.layers.Dense(26, activation='softmax'),   # here 20 is number of neurons
    keras.layers.Dense(15, activation='softmax'),
    keras.layers.Dense(1, activation='sigmoid'),   # output layer
])

    model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

    if weights==-1:
        model.fit(x_train,y_train,epochs=100)
    else:
        model.fit(x_train,y_train, epochs=100, class_weight=weights)
              
    print(model.evaluate(x_test,y_test))
    y_preds = model.predict(x_test)
    y_preds = np.round(y_preds)     #rounding off to nearest integer
    
    

    print('Classification Report:', classification_report(y_test,y_preds))

    return y_preds
             
```

# METHOD - UNDERSAMPLING
In this method we will reduce the instance from majority class to balance the class distribution

```python
# Class count
count_class_0, count_class_1 = df1.Churn.value_counts()

# Divide by class
df_class_0 = df2[df2['Churn'] == 0]
df_class_1 = df2[df2['Churn'] == 1]
```

```python
# Undersample 0-class and concat the DataFrames of both class
df_class_0_under = df_class_0.sample(count_class_1)           # Randomly selects a subset of instances from the majority class (df_class_0) to match the number of instances in the minority class (count_class_1)
df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)

print('Random under-sampling:')
print(df_test_under.Churn.value_counts())
```

```python
x = df_test_under.drop('Churn',axis='columns')
y = df_test_under['Churn']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=15, stratify=y)
```

```python
# Number of classes in training Data
y_train.value_counts()
```

```python
y_preds = ANN(x_train, y_train, x_test, y_test, 'binary_crossentropy', -1)
```
